```python
import numpy as np
def sigmoid(x):
    return 1/(1+ np.exp(-x))
def d_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
x =np.linspace(-10,10,100)
y = sigmoid(x)
d = d_sigmoid(x)
```

```python
import pandas as pd
df = pd.DataFrame ({"x":x,"sigmoid(X)":y,"d_sigmoid(x)":d})
df
```

```python
import matplotlib.pyplot as plt
plt.style.use("dark_background")
fig = plt.figure(figsize=(18,10))
plt.plot(x,y,c="lightgreen",linewidth = 4.0,label="$\sigma(x)$")
plt.plot(x,d,c="lightblue",linewidth=4.0,label="$\\frac{d}{dx} \sigma(x)$")
plt.legend(prop={'size':20})
plt.show()
```

```python
from sklearn .linear_model import  LogisticRegression
```

```python
hours_spent_learning = np.linspace(1,100,100)
weights = hours_spent_learning / len (hours_spent_learning)
pass_test = np.random.binomial(1,weights)
```

```python
print('First Ten Values')
print('\thours_spent_learning:\t',hours_spent_learning[:10])
print('\tweights:\t\t',weights[:10])
print('\tpass_test:\t\t',pass_test[:10])
```

```python
x = hours_spent_learning.reshape(-1,1)
y =pass_test
model= LogisticRegression()
model.fit(x,y)
```

```python
pred = sigmoid(x*model.coef_[0]+model.intercept_[0])
fig = plt.figure(figsize=(15,9))
plt.plot(x,pred,c="lightblue",linewidth=4.0)
plt.scatter(
x[(y==1).ravel()],
y[(y==1).ravel()],
marker =".",
c="lightgreen",
linewidth = 1.0,
label ="passed",)
plt.scatter(
x[(y==0).ravel()],  
y[(y==0).ravel()],
marker =".", 
linewidth=1.0,
label="failled", )
plt.axhline(y=0.5,color='red',linestyle='--',label="boundary")
plt.xlabel("hours_spent_learning")
plt.ylabel('p("passing the test")')
plt.legend(frameon=False,loc="best",bbox_to_anchor=(0.5,0.0,0.5,0.5),prop={'size':20})
plt.show()
```
